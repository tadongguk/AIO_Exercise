```python

new_dataset = 'module3/week1/opsd_germany_daily.csv'
data = pd.read_csv(new_dataset)
data
```

```python
opsd_daily = data.set_index('Date')
opsd_daily
```

```python
opsd_daily = pd.read_csv(new_dataset, index_col=0, parse_dates=True)

opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday'] = opsd_daily.index.weekday
opsd_daily.sample(5, random_state=0)
```

```python
opsd_daily.loc['2012-08-10':'2012-08-11']
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11, 9)})
opsd_daily['Consumption'].plot(linewidth=0.5)
```

```python
cols_plot = ['Consumption', 'Solar', 'Wind']
axes = opsd_daily[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily totals')
plt.show()
```

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head(3)
```

```python
print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

```python
start, end = '2017-01', '2017-06'
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth=0.5, label='Weekly')
ax.set(title='Solar Energy Consumption (2017)')
ax.set_ylabel('Solar Energy (GWh)')
ax.legend()
plt.show()
```
